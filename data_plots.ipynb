```python
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
signal = np.load("data/data_signal_B_1ch_0000.npy").astype(np.float32)
noise = np.load("data/data_noise_3.0SNR_1ch_0001.npy").astype(np.float32)

#noise = np.zeros((1000000, 1, 256), dtype=np.float32)
#for i in range(10):
#    noise[(i) * 100000:(i + 1) * 100000] = np.load(f"data/data_noise_3.0SNR_1ch_{i:04d}.npy").astype(np.float32)

#print(noise.shape)

l = len(noise[0,0,:])
x = np.arange(1,l+1)
```

```python
#ID plots
for i in range(100):
    plt.plot(x, noise[i*1000,0,:])
    plt.xlabel('Time [ns]')
    plt.ylabel('Amplitude [V]')
    plt.savefig('plots/noise_'+str(i)+'.png')
    plt.clf()

    plt.plot(x, signal[i*1000,0,:])
    plt.xlabel('Time [ns]')
    plt.ylabel('Amplitude [V]')
    plt.savefig('plots/signal_'+str(i)+'.png')
    plt.clf()

```
